```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
```

```python
sns.boxplot(x ="timepoint",
y ="signal",
hue ="region",
data = fmri)
plt.show()

#hue="region": This parameter adds another dimension to the plot by using the "region" variable to group the data.
#It creates separate boxes for each unique value of "region", allowing for comparison between regions within each timepoint.
```
